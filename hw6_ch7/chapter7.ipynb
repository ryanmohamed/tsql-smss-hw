1\. Sum the running total for each order for each employee.Â 

1b. Sum the running total of this and the past 2 orders for each employees orders.

1c. Sum the running total of this and the following 2 orders for each employees orders.

```sql
USE TSQLV4;

-- running total
SELECT empid, ordermonth, val,
--window aggregate
--performs calc on a set of rows
  SUM(val) OVER(PARTITION BY empid
                ORDER BY ordermonth
                --partition default
                --grows for each new calc
                --unbounded means the last possible row
                ROWS BETWEEN unbounded PRECEDING 
                         AND CURRENT ROW) AS runval
FROM Sales.EmpOrders;

-- running total of past 2 orders for each order
SELECT empid, ordermonth, val,
--window aggregate
--performs calc on a set of rows
  SUM(val) OVER(PARTITION BY empid
                ORDER BY ordermonth
                --partition default
                --grows for each new calc
                ROWS BETWEEN 2 PRECEDING 
                         AND CURRENT ROW) AS runval
FROM Sales.EmpOrders;

-- running total of next 2 orders for each order
-- keep in mind if there is no next two, we get back nothing
SELECT empid, ordermonth, val,
--window aggregate
--performs calc on a set of rows
  SUM(val) OVER(PARTITION BY empid
                ORDER BY ordermonth
                --partition default
                --grows for each new calc
                ROWS BETWEEN CURRENT ROW
                         AND 2 FOLLOWING) AS runval
FROM Sales.EmpOrders;
```

```sql
use TSQLV4;

SELECT orderid, custid, val,
  ROW_NUMBER() OVER(ORDER BY val) AS rownum,
  RANK()       OVER(ORDER BY val) AS rank,
  DENSE_RANK() OVER(ORDER BY val) AS dense_rank,
    --rows are made based off of the value of val for the current row
    --for a tie, we can use dense_rank to give orders with the same val, the same rank
    --rather than use a tiebreaker
  NTILE(10)    OVER(ORDER BY val) AS ntile
    --divides the ordered rows into 10 "rougly" equal parititons
FROM Sales.OrderValues
ORDER BY val;
```

3\. Partition each customers orders and rank based on val.

```sql
use TSQLV4;

SELECT orderid, custid, val,
    --think of partition as a box based on some commonality
  ROW_NUMBER() OVER(PARTITION BY custid
                    ORDER BY val) AS rownum
FROM Sales.OrderValues
ORDER BY custid, val;
```

```sql
use TSQLV4;

--creates a rank for each unique order value 
SELECT DISTINCT val, ROW_NUMBER() OVER(ORDER BY val) AS rownum
FROM Sales.OrderValues;
```

```sql
use TSQLV4;

SELECT val, ROW_NUMBER() OVER(ORDER BY val) AS rownum
FROM Sales.OrderValues
GROUP BY val; --removes duplicates in the same manner without DISTINCT
```

```sql
use TSQLV4;

--returns null rather than empty cell
SELECT custid, orderid, val,
    --get last val based on partition 
    --in this case (if custid has no next or lag - NULL)
  LAG(val)  OVER(PARTITION BY custid
                 ORDER BY orderdate, orderid) AS prevval,
    --get next val based on partition
    --in this case (if custid has no next or lag - NULL)
  LEAD(val) OVER(PARTITION BY custid
                 ORDER BY orderdate, orderid) AS nextval
FROM Sales.OrderValues
ORDER BY custid, orderdate, orderid;
```

```sql
use TSQLV4;

SELECT custid, orderid, val,
    --for the custids running amount of rows ranks on orderdate and orderid
    --get the first value
  FIRST_VALUE(val) OVER(PARTITION BY custid
                        ORDER BY orderdate, orderid
                        ROWS BETWEEN UNBOUNDED PRECEDING
                                 AND CURRENT ROW) AS firstval,
    --for the custids running amount of rows ranks on orderdate and orderid
    --get the first value
  LAST_VALUE(val)  OVER(PARTITION BY custid
                        ORDER BY orderdate, orderid
                        ROWS BETWEEN CURRENT ROW
                                 AND UNBOUNDED FOLLOWING) AS lastval
FROM Sales.OrderValues
ORDER BY custid, orderdate, orderid;
```

```sql
use TSQLV4;

SELECT orderid, custid, val,
    --notice how there is no partition for this one
    --sum of ALL values, no need for group by
    --all rows will share values (very powerful)
  SUM(val) OVER() AS totalvalue,
    --partitioned based on customer
    --sum of val for CUSTID only
  SUM(val) OVER(PARTITION BY custid) AS custtotalvalue
FROM Sales.OrderValues;
```

Get the percentage the current order takes up for ALL ORDERS and ALL ORDERS for that customer.

```sql
use TSQLV4;

SELECT orderid, custid, val,
  100. * val / SUM(val) OVER() AS pctall,
  100. * val / SUM(val) OVER(PARTITION BY custid) AS pctcust
FROM Sales.OrderValues;
```

Sum orders quantity between an employee and each customer, for each employee.

```sql
use TSQLV4;

-- this pivots the above table
-- pivoting on empid (the corner at which the table rotates)

select empid,
    --perform aggregate sum of qty against specifc customer ids, alias as custID
    --if no orders exist with custid, NULL
    sum(case when custid = 'A' then qty end) as A,
    sum(case when custid = 'B' then qty end) as B,
    sum(case when custid = 'C' then qty end) as C,
    sum(case when custid = 'D' then qty end) as D
from dbo.Orders
group by empid
```

```sql
use TSQLV4;

select empid, A, B, C, D
from ( select empid, custid, qty --very important from doesn't contain columns we dont want
    from dbo.Orders 
    ) as D
    --first argument is our aggregate and column to aggregate
    --for pivot columns(s)
    --in(pivot values (the unique values from pivot column to pivot upwards))
    --think that usually this column has DOWNWARD rows (provided by a group by so uniqueness is there)
    --we want to PIVOT these column so these specific rows become columns
    --in provides the aliases for this P
    pivot(sum(qty) for custid in(A, B, C, D)) as P
```

```sql
use TSQLV4;

--pivots use implicit group bys!

select empid, A, B, C, D
from dbo.Orders --without specifying columns from orders, its column are used
                --in an implicit group by!
                --so our unique employee rows are ruined
    pivot(SUM(qty) for custid IN(A, B, C, D)) as P

--with explicit group by, the equivalent would be
SELECT empid,
  SUM(CASE WHEN custid = 'A' THEN qty END) AS A,
  SUM(CASE WHEN custid = 'B' THEN qty END) AS B,
  SUM(CASE WHEN custid = 'C' THEN qty END) AS C,
  SUM(CASE WHEN custid = 'D' THEN qty END) AS D  
FROM dbo.Orders
GROUP BY orderid, orderdate, empid;
```

```sql
use TSQLV4

--now the reverse, get sum of qtys for each unique customers
--amongst all employees

select custid, [1], [2], [3]
from (select empid, custid, qty
    from dbo.Orders) as D
    pivot(sum(qty) for empid in([1], [2], [3])) as P
```

```sql
USE TSQLV4;

DROP TABLE IF EXISTS dbo.EmpCustOrders;

CREATE TABLE dbo.EmpCustOrders
(
  empid INT NOT NULL
    CONSTRAINT PK_EmpCustOrders PRIMARY KEY,
  A VARCHAR(5) NULL,
  B VARCHAR(5) NULL,
  C VARCHAR(5) NULL,
  D VARCHAR(5) NULL
);

INSERT INTO dbo.EmpCustOrders(empid, A, B, C, D)
  SELECT empid, A, B, C, D
  FROM (SELECT empid, custid, qty
        FROM dbo.Orders) AS D
    PIVOT(SUM(qty) FOR custid IN(A, B, C, D)) AS P;

SELECT * FROM dbo.EmpCustOrders;
```

Grouping Sets

```sql
use TSQLV4;

--sum of each unique customer-employee order
select empid, custid, sum(qty) as sumqty
from dbo.Orders
group by empid, custid

--sum of each unique employee order (amongst all customers)
select empid, sum(qty) as sumqty
from dbo.Orders
group by empid

--sum of each unique customer order (amongst all employees)
select custid, sum(qty) as sumqty
from dbo.Orders
group by custid

--sum of all orders irrespecive to any column
select sum(qty) as sumqty
from dbo.Orders

--how can we join these different grouping set into one table?
--obviously set operators
--very easy to interpret

--sum of each unique customer-employee order
--first designates columns for the rest, assumes equal number of columns
select empid, custid, sum(qty) as sumqty
from dbo.Orders
group by empid, custid

UNION ALL

--sum of each unique employee order (amongst all customers)
select empid, null, sum(qty) as sumqty
from dbo.Orders
group by empid

UNION ALL

--sum of each unique customer order (amongst all employees)
select null, custid, sum(qty) as sumqty
from dbo.Orders
group by custid

UNION ALL

--sum of all orders irrespecive to any column
select null, null, sum(qty) as sumqty
from dbo.Orders
```

Grouping sets subclause.

```sql
use TSQLV4

--obviously the above method is code intensive
--therefore the need for its own clause

select empid, custid, sum(qty) as sumqty
from dbo.Orders
group by
    grouping sets ( --multiple groups
        (empid, custid), --unique pair
        (empid),
        (custid),
        ()
    )
```

```sql
use TSQLV4;
--cube subclause creates all combinations of nonaggregate columns
select empid, custid, sum(qty) as sumqty
from dbo.Orders
group by cube(empid, custid)
```

Rollup subclause.

```sql
use TSQLV4;

select year(orderdate) as orderyr,
    month(orderdate) as  ordermonth,
    day(orderdate) as orderday,
    sum(qty) as sumqty
from dbo.Orders
group by rollup(year(orderdate), month(orderdate), day(orderdate))
```

```sql
use TSQLV4;

SELECT empid, custid, SUM(qty) AS sumqty
FROM dbo.Orders
GROUP BY CUBE(empid, custid);

SELECT
  GROUPING(empid) AS grpemp,
  GROUPING(custid) AS grpcust,
  empid, custid, SUM(qty) AS sumqty
FROM dbo.Orders
GROUP BY CUBE(empid, custid);

SELECT
  GROUPING_ID(empid, custid) AS groupingset,
  empid, custid, SUM(qty) AS sumqty
FROM dbo.Orders
GROUP BY CUBE(empid, custid);
```
